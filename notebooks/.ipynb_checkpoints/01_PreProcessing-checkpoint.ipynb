```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
FUNDUS_PATH = '../data/image001.png'
```

```python
image = cv2.imread(FUNDUS_PATH)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
print('size: {} shape: {}'.format(image.size, image.shape))
```

```python
#gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(image)
plt.show()
```

```python
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
image.shape
```

```python
# Extract the green channel
```

```python
green = image[:,:,1]
```

```python
plt.imshow(green, cmap = plt.cm.gray)
plt.show()
```

```python
# CLAHE
```

```python
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
cl1 = clahe.apply(green)
```

```python
plt.imshow(cl1, cmap = plt.cm.gray)
plt.show()
```

```python
# Image Denoising
```

```python
dst = cv2.fastNlMeansDenoising(cl1)
```

```python
plt.imshow(dst, cmap = plt.cm.gray)
plt.show()
```

```python
# Another Denoising according to paper
```

```python
kernel = np.ones((9,9),np.float32)/81
dst = cv2.filter2D(cl1,-1,kernel)
```

```python
plt.imshow(dst, cmap = plt.cm.gray)
plt.show()
```

```python
# Background Exclusion
```

```python
cv2_subt = cv2.subtract(dst, image[:,:,1])
```

```python
plt.imshow(cv2_subt > 0 , cmap = plt.cm.gray)
plt.show()
```
